```python
# Recursion: Invoking a function from within the function itself
```

```python
# the function can repeat itself 3000 times before it hits a error (memory limit hit)
import sys
sys.getrecursionlimit()
```

```python
# let's test out if that is true.

# we hit an error at 2958 which is approx 3000.

def increment(num):
    
    print(num, end = " ")
    
    increment(num + 1)
    return
```

```python
increment(1)
```

```python
# we can change the memory limit in recursion

sys.setrecursionlimit(100)
```

```python
increment(1)
```

```python
# Recursion Condition needs to set to get out of a recursion. Otherwise we get stuck in an infinite loop/recursion error

def decrement(num):
    
    if num == 0:  # terminating condition is set, when the num == 0 the function will terminate
        return
    
    print(num)
    decrement(num - 1)
```

```python
decrement(10)
```

```python
def recursive_sum(num):
    
    if num == 0:  # when the num == 0 the function will terminate
        return 0
    
    result = num + recursive_sum(num - 1)
    
    return result

# result = 5 + recursive_sum(4+3+2+1)
```

```python
recursive_sum(5)
```

```python
# using recursion to calculate factorial of a number

def factorial(number):
    
    # error check condition first
    if number < 0:
        print("Sorry, factorial does not exists for negative numbers")
        return
        
    if number == 0:
        return 1
    
    return number * factorial(number - 1)
```

```python
factorial(3)
```

```python
def fibonacci(number, fib_series):
    
    if number < 2:
        return
    
    l = len(fib_series)
    
    new_number = fib_series[l - 1] + fib_series[l - 2]
    
    fib_series.append(new_number)
    print("Series so far", fib_series)
    
    fibonacci(number - 1, fib_series)
```

```python
# example: we are passing a list of numbers with 2 elements and we want a series of 4 fib numbers returned

series = [0, 1]

fibonacci(4, series)
```
